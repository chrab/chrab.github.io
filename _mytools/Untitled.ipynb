```python
import re

def extract_data2(s):
    pattern = re.compile(r"""\|\s*                 # opening bar and whitespace
                             '(?P<name>.*?)'       # quoted name
                             \s*\|\s*(?P<n1>.*?)   # whitespace, next bar, n1
                             \s*\|\s*(?P<n2>.*?)   # whitespace, next bar, n2
                             \s*\|""", re.VERBOSE)
    match = pattern.match(s)

    name = match.group("name")
    n1 = float(match.group("n1"))
    n2 = float(match.group("n2"))

    return (name, n1, n2)

f=open("talks.txt")

fyaml=open("talks.yml",mode="w")

for line in f:
  txt=line.strip()
  if txt != "":    
    out=re.search(r'resumeitem\{(.*?)\}',txt).group(1).split(",")
    
    date=out[0]
    location=out[1]
    
    title=re.search(r'textit\{(.*?)\}',txt).group(1)
    
    text=re.search(r'\.\}(.*?)\,',txt)
    print(text)
    if text is None:      
      text=re.search(r'\}\,(.*?)\,',txt)      
    if text is not None:
      text=text.group(1)      
      text=text.replace(r"\mbox{","")
      text=text.replace(r"}","")
      text=text.replace(r"\newline","")
    else:
      print(txt)
      
    
    
    typ=txt.split(",")[-1]
    typ=typ.replace(r"\textbf{invited}","invited")
    typ=typ.replace(r"}","")
    typ=typ.replace(r"\\","")
    typ=typ.replace(r".","")
    typ=typ.strip()
    if typ=="invited":
      typ="Invited Talk"
    elif typ=="contributed":
      typ="Contributed Talk"
       
    fyaml.write("- title: \""+title.strip()+"\"\n")
    fyaml.write("  date: "+date.strip()+"\n")
    fyaml.write("  location: "+location.strip()+"\n")
    fyaml.write("  type: "+typ.strip()+"\n")
    fyaml.write("  text: \""+text.strip()+"\"\n")
    fyaml.write("\n")
    

fyaml.close()
f.close()
```
